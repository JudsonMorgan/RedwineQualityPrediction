### Modeling and Evaluation

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
updated_data.head()
```

```python
X = updated_data.iloc[:,:-1]
y = updated_data.iloc[:, -1]
```

```python
X.head()
```

```python
y.head()
```

```python
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
X_scaled
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
reg = LogisticRegression()
reg.fit(X_train, y_train)
```

- To ensure that our model don't overfit or underfit; we will be checking the accuracy score of both the test and training data

```python
reg.score(X_test, y_test)
```

```python
reg.score(X_train, y_train)
```

```python
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
```

```python
rfc.score(X_test, y_test)
```

```python
rfc.score(X_train, y_train)
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
```

```python
dtc.score(X_test, y_test)
```

```python
dtc.score(X_train, y_train)
```

```python
Lsvc = LinearSVC()
Lsvc.fit(X_train, y_train)
```

```python
Lsvc.score(X_test, y_test)
```

```python
Lsvc.score(X_train, y_train)
```

For our model, The Logistic Regression is the better model has it does not overfit or underfit on our data.

```python
y_pred = reg.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
plt.figure(figsize=(10,10))
plt.scatter(y_test, y_pred, c='crimson')
plt.yscale('log')
plt.xscale('log')

p1 = max(max(y_pred), max(y_test))
p2 = min(min(y_pred), min(y_test))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('True Values', fontsize=15)
plt.ylabel('Predictions', fontsize=15)
plt.axis('equal')
plt.show()
```

### Hyperparamter tuning

```python
np.logspace(-3,3,7)
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

grid = {'penalty': ['l1', 'l2'],
           'C': np.logspace(-3,3,7),
            'solver':['newton-cg', 'lbfgs', 'liblinear']}

reg_grid = GridSearchCV(reg,
                       param_grid = grid,
                       n_jobs=-1,
                       cv=10,
                       verbose=True,
                       return_train_score=True)
reg_grid.fit(X_train, y_train)
```

```python
reg_grid.score(X_test, y_test)
```

```python
reg_grid.score(X_train, y_train)
```

```python
reg_grid.best_params_
```

```python
reg_grid.best_score_
```

```python
reg_grid.cv_results_
```

### Using RandomizedSearchCV

```python
parameters = {'penalty': ['l1', 'l2'],
             'C': np.logspace(-3,3,7),
             'solver': ['newton-cg', 'lbfgs', 'liblinear']}

reg_param = RandomizedSearchCV(reg,
                              param_distributions = parameters,
                              verbose=True,
                              cv=10,
                              random_state=42,
                              return_train_score=True)
reg_param.fit(X_train, y_train)
```

```python
reg_param.score(X_test, y_test)
```

```python
reg_param.score(X_train, y_train)
```
